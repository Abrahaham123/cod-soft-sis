```python




calificacion = float(input("Introduce la calificación: "))

if calificacion > 90:

    print("va a Grupo A.")

elif calificacion >= 80 and calificacion <= 90: # NO "&&" en Python

    print("va a Grupo B.") 

elif calificacion >= 60 and calificacion <= 80:

    print("→ Accede a Grupo C.")

elif calificacion < 60:

    print("va  a Grupo D.")

else:

    print("lacalificación no válida.")



```

```python





precio = float(input("Introduce el precio de la bicicleta: "))







A = precio * 0.15

B = precio * 0.10

C = precio * 0.05






if precio > 100000:

    print('Aplica Impuesto de 15%, son $', A)

elif precio >= 50000 and precio <= 100000:

    print('Aplica Impuesto de 10%, son $', B)

elif precio < 50000:

    print('Aplica Impuesto de 5%, son $', C)
```

```python






entrada = int(input('Ingresa el numero de la semana: '))






if entrada == 1:

    print('LUNES')

elif entrada == 2:

    print('MARTES')

elif entrada == 3:

    print('MIERCOLES')

elif entrada == 4:

    print('JUEVES')

elif entrada == 5:

    print('VIERNES')

elif entrada == 6:

    print('SABADO')

elif entrada == 7:

    print('DOMINGO')

else:

    print('Selecciona un dia')






if entrada > 7 or entrada < 1:

    print('No entradara ', entrada, " dentradaro del rango")

    print('Reinicia e intentradaa de nuevo')


```
